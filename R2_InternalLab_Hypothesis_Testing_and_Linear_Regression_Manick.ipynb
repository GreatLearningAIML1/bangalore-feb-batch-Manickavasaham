## Question 4 

*In one or two sentences explain about **TypeI** and **TypeII** errors.*

#### Answer:

```python
A type I error occurs when the null hypothesis (H0) is true, should not be rejected. But is rejected. 
Type 1 error is a "false alarm" The probability of a Type 1 error occuring is 'α'
```

```python
A type II error occurs when the null hypothesis is false, but erroneously fails to be rejected.  
A Type II error  represents missed opportunity and is committed when we fail to believe a true alternative hypothesis.
the prob. of Type II error occuring is β  
```

# Question 7 
Plot the distribution created in the above question.

```python
rv = binom(n, p)
fig, ax = plt.subplots(1, 1)
ax.vlines(k, 0, rv.pmf(k), colors='k', linestyles='-', lw=1,
      label='pmf')
ax.legend(loc='best', frameon=False)
plt.show()
```
